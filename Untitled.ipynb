# Income prediction

```python
#importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Removing Warnings

import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('dataset.csv')
```

```python
df.head()
```

```python
# Removing missing value '?'

df=df[df['workclass']!='?']
```

```python
df.head()
```

```python
df.info()
```

```python
#Checking if there is still missing values '?'

categorical_data=df.select_dtypes(include=['object'])
categorical_data.apply(lambda x: x=='?',axis=0).sum()
```

```python
df=df[df['occupation']!='?']
df=df[df['native.country']!='?']
```

```python
df.head()
```

```python
df.info()
```

```python
#Getting all categorical data

categorical_data=df.select_dtypes(include=['object'])
categorical_data.head()
```

```python
#importing LabelEncoder to convert the categorical data to numerical data

from sklearn.preprocessing import LabelEncoder
```

```python
l_enc=LabelEncoder()
```

```python
#converting categorical data to numerical data

categorical_data=categorical_data.apply(l_enc.fit_transform)
```

```python
#converted data

categorical_data.head()
```

```python
#deleting categorical data from original data

df=df.drop(df.select_dtypes(include=['object']).columns,axis=1)
```

```python
#concatening numerical data with converted categorical data

df=pd.concat([df,categorical_data],axis=1)
```

```python
df.head()
```

```python
df.info()
```

```python
df['income']=df['income'].astype('category')
```

```python
X=df.drop('income',axis=1)
y=df['income']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30)
```

```python
X.shape
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
d_tree=DecisionTreeClassifier(max_depth=5)
```

```python
d_tree.fit(X_train,y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
prediction=d_tree.predict(X_test)
```

```python
confusion_matrix(y_test,prediction)
```

```python
accuracy_score(y_test,prediction)
```

```python
print(classification_report(y_test,prediction))
```
